```python
import numpy as np

age = [a/50*34+3 for a in list(range(50))]

asd = np.linspace(4.8, 4.75)
hra = np.linspace(2.75, 3.28)
lrc = np.linspace(1.2, 1.45)

asd_std = np.linspace(-0.2, 0.2)**2+0.2
hra_std = np.linspace(-0.1, 0.3)**2+0.15
lrc_std = np.linspace(-0.1, 0.2)**2+0.05

corr_x = [3, 6, 9, 12, 19, 24, 36]
corr_y = [4.3, 5.1, 4.3, 3.4, 5.1, 4.9, 5.0]

asd_meas = 4.5
hra_meas = 2.0
lrc_meas = 1.5
```

```python
## Step 1
import matplotlib.pyplot as plt
%matplotlib inline

from matplotlib import style
style.use('ggplot') 


plt.figure(figsize=(12,15))

plt.plot(age, asd)
plt.plot(age, hra)
plt.plot(age, lrc) 

plt.plot(corr_x, corr_y)

plt.axhline(asd_meas, 0, 40)
plt.axhline(hra_meas, 0, 40)
plt.axhline(lrc_meas, 0, 40)
```

```python
# Step 2: add labels
plt.figure(figsize=(12,15))

plt.plot(age, asd)
plt.plot(age, hra)
plt.plot(age, lrc) 

plt.plot(corr_x, corr_y)

plt.axhline(asd_meas, 0, 40)
plt.axhline(hra_meas, 0, 40)
plt.axhline(lrc_meas, 0, 40)

plt.xlabel('Age (months)')
plt.ylabel('SVM predicted score')
plt.title('Score based on distance from SVM Plane')
```

```python
# Step3: give everything the right colour
plt.figure(figsize=(12,15))


plt.plot(age, asd, 'r')
plt.plot(age, hra, 'g')
plt.plot(age, lrc, 'b') 


plt.plot(corr_x, corr_y, 'X-k', label='Correlation', markersize=18)

plt.axhline(asd_meas, 0, 40, color = 'r', linestyle = '--')
plt.axhline(hra_meas, 0, 40, color = 'g', linestyle = '--')
plt.axhline(lrc_meas, 0, 40, color = 'b', linestyle = '--')

plt.xlabel('Age (months)')
plt.ylabel('SVM predicted score')
plt.title('Score based on distance from SVM Plane')
```

```python
### step 4
plt.plot(age, asd, 'r')
plt.fill_between(age, asd-asd_std, asd+asd_std,
    alpha=0.2,  facecolor='r')
plt.plot(age, hra, 'g')
plt.fill_between(age, hra-hra_std, hra+hra_std,
    alpha=0.2,  facecolor='g')
plt.plot(age, lrc, 'b') 
plt.fill_between(age, lrc-lrc_std, lrc+lrc_std,
    alpha=0.2,  facecolor='b')


plt.plot(corr_x, corr_y, 'X-k', label='Correlation', markersize=18)

plt.axhline(asd_meas, 0, 40, color = 'r', linestyle = '--')
plt.axhline(hra_meas, 0, 40, color = 'g', linestyle = '--')
plt.axhline(lrc_meas, 0, 40, color = 'b', linestyle = '--')

plt.xlabel('Age (months)')
plt.ylabel('SVM predicted score')
plt.title('Score based on distance from SVM Plane')
```

```python
# Step 5 Add text
plt.figure(figsize=(12,15))

plt.plot(age, asd, 'r')
plt.fill_between(age, asd-asd_std, asd+asd_std,
    alpha=0.2,  facecolor='r')
plt.plot(age, hra, 'g')
plt.fill_between(age, hra-hra_std, hra+hra_std,
    alpha=0.2,  facecolor='g')
plt.plot(age, lrc, 'b') 
plt.fill_between(age, lrc-lrc_std, lrc+lrc_std,
    alpha=0.2,  facecolor='b')


plt.plot(corr_x, corr_y, 'X-k', label='Correlation', markersize=18)

plt.axhline(asd_meas, 0, 40, color = 'r', linestyle = '--')
plt.axhline(hra_meas, 0, 40, color = 'g', linestyle = '--')
plt.axhline(lrc_meas, 0, 40, color = 'b', linestyle = '--')

plt.text(13, 4.51, 'Measured CSS score: ASD')
plt.text(13, 2.01, 'Measured CSS score: HRA-')
plt.text(13, 1.51, 'Measured CSS score: LRC-')

plt.text(26, 4.7, 'Predicted CSS score: ASD')
plt.text(26, 3.05, 'Predicted CSS score: HRA-')
plt.text(26, 1.3, 'Predicted CSS score: LRC-')

plt.text(8, 3.7, 'Correlation x 10 -1')


plt.xlabel('Age (months)')
plt.ylabel('SVM predicted score')
plt.title('Score based on distance from SVM Plane')
```

```python
# Step 5 Add text
plt.figure(figsize=(12,15))

plt.plot(age, asd, 'r', label='asd')
plt.fill_between(age, asd-asd_std, asd+asd_std,
    alpha=0.2,  facecolor='r', label='asd std')
plt.plot(age, hra, 'g', label='hra-')
plt.fill_between(age, hra-hra_std, hra+hra_std,
    alpha=0.2,  facecolor='g', label = 'hra- std')
plt.plot(age, lrc, 'b', label='lrc-') 
plt.fill_between(age, lrc-lrc_std, lrc+lrc_std,
    alpha=0.2,  facecolor='b', label='lrc- std')


plt.plot(corr_x, corr_y, 'X-k', label='Correlation', markersize=18)

plt.axhline(asd_meas, 0, 40, color = 'r', linestyle = '--', label='measured asd')
plt.axhline(hra_meas, 0, 40, color = 'g', linestyle = '--', label='measured hra-')
plt.axhline(lrc_meas, 0, 40, color = 'b', linestyle = '--', label='measured lrc-')

plt.text(13, 4.51, 'Measured CSS score: ASD')
plt.text(13, 2.01, 'Measured CSS score: HRA-')
plt.text(13, 1.51, 'Measured CSS score: LRC-')

plt.text(26, 4.7, 'Predicted CSS score: ASD')
plt.text(26, 3.05, 'Predicted CSS score: HRA-')
plt.text(26, 1.3, 'Predicted CSS score: LRC-')

plt.text(8, 3.7, 'Correlation x 10 -1')

plt.legend(bbox_to_anchor=(1.20, .5))

plt.xlabel('Age (months)')
plt.ylabel('SVM predicted score')
plt.title('Score based on distance from SVM Plane')

plt.savefig('example_figures/replica_dori', dpi=500)
```
