Answer exercise 2: leaders

```python
import random
all_numbers = [random.randint(0,50) for i in range(5)]
leaders = []
i = 0
for no in all_numbers[:-1]:
    if no > sum(all_numbers[i+1:]):
        leaders.append(no)
    i += 1
```

```python
all_numbers = [random.randint(0,50) for i in range(5)]
leaders = []
for i,no in enumerate(all_numbers[:-1]):
    if no > sum(all_numbers[i+1:]):
        leaders.append(no)
```

```python
print(all_numbers)
print(leaders)
```

Answer exercise 3: find spikes in calcium data

```python
import numpy as np
calciumdata = np.random.exponential(1,(100,1))
threshold = np.nanmean(calciumdata) + 3*np.nanstd(calciumdata)

def find_spikes(data, threshold = 0.9):
    thresholded_data = np.copy(data) # if you don't do np.copy here, the thresholded data will overwrite the original data
    thresholded_data[data < threshold] = 0
    spikestimes = [i for i, d in enumerate(data) if d > threshold]
    return thresholded_data, spikestimes
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
def test_find_spikes(data, threshold):
    thresh_data, spike_times = find_spikes(data, threshold)
    plt.plot(data)
    plt.plot(thresh_data)
    plt.scatter(spike_times, np.asarray(data)[spike_times])
    plt.legend(('Original data', 'Thresholded data', 'Spike Times'))
test_find_spikes(calciumdata, threshold)
```
